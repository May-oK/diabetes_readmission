```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("../Output/OPT_Training_Set_grouped_binary.csv")
test = pd.read_csv("../Output/OPT_Testing_Set_grouped_binary.csv")
```

```python
train = train.drop("Unnamed: 0", axis=1)
test = test.drop("Unnamed: 0", axis=1)
```

```python
treatments = ["med_change_1_NoMed","med_change_1_Steady", "med_change_1_Up", "med_change_2_Mixed",
                    "med_change_2_NoMed", "med_change_2_Steady", "med_change_2_Up",
                    "med_change_3_NoMed", "med_change_3_Steady", "med_change_3_Up", 
                    "med_change_4_NoMed", "med_change_4_Steady", "med_change_4_Up",
                    "med_change_5_NoMed", "med_change_5_Steady", "med_change_5_Up",
                    "med_change_6_NoMed", "med_change_6_Steady", "med_change_6_Up",
                    "med_change_other_Yes"]
```

```python
treatments_train = train[treatments]
treatments_test = test[treatments]
treatments_train
```

```python
unique_counts = treatments_train.groupby(list(treatments_train.columns)).size().reset_index(name='Counts')
unique_counts_sorted = unique_counts.sort_values(by=["Counts"], ascending=False)
unique_counts_sorted
```

### Extract rows with top 3 treatments only

```python
unique_counts_sorted_top3 = unique_counts_sorted.head(3)
sum(unique_counts_sorted_top3['Counts'])
```

```python
Testing_Filtered_top3 = pd.merge(test, unique_counts_sorted_top3.iloc[:, 0:20], how="inner", on=list(treatments_train.columns))
Testing_Filtered_top3
```

```python
Training_Filtered_top3 = pd.merge(train, unique_counts_sorted_top3.iloc[:, 0:20], how="inner", on=list(treatments_train.columns))
Training_Filtered_top3
```

```python
#Save CSVs
Training_Filtered_top3.to_csv("../Output/Training_Filtered_top3_long.csv")
Testing_Filtered_top3.to_csv("../Output/Testing_Filtered_top3_long.csv")
```

### Extract rows with top 10 treatments only

```python
## Extract top 10 treatments
unique_counts_sorted_top10 = unique_counts_sorted.head(10)
sum(unique_counts_sorted_top10['Counts'])
```
